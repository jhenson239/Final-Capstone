```python
# Imports
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\brand\OneDrive\Documents\Bootcamp Projects\Capstone Project 2\Cleaned Data\cleaned_data")
```

```python
df = df.loc[:, df.columns != 'Unnamed: 0']
df.head(5)
```

```python
# Basic summarization of data.
df2 = df.drop(["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"], axis=1)
df2.head()
df2.describe()
```

```python
df2.sample(n=5)
```

```python
# heatmap
plt.figure(figsize=(10,5))
columns_to_include = ['Year_of_Release','Global_Sales','Critic_Score','Critic_Count',
                      'User_Score','User_Count']
df_heat = df2[columns_to_include]
sns.heatmap(data= df_heat.corr(), annot=True, cmap='coolwarm',linewidth=.5)

# Not noting any strong correlation between any variables or features. 
```

```python
sns.pairplot(df2, hue='Global_Sales', height=2, aspect=2)
```

```python
# The next 6 cells are generating decriptive discriptive statistics for differnts groups. Really interested in sales performance for each group. 
# The PLaystion platorm is the strongest, PS1,PS2,PS2,"ect", including the PSP and PSV. Surpring considering nintendoes market share by publisher.
platform_descriptions = df2.groupby('Platform')
for platform,group in platform_descriptions:
    print(f"Descriptive statistics for Platfroms: {platform}")
    print(group.describe())
    print("\n")

grouped_sales = df2.groupby('Platform')["Global_Sales"].sum()
print(f"Total gloabl Sales by {grouped_sales}")

plt.figure(figsize=(10, 6))
grouped_sales.plot(kind='bar', color='skyblue')
plt.title('Total Global Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('Total Global Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
#Grouping by genre. Action has a firm lead followed by shooter. Makes since considering the Halo and Call of Duty Franchises.
#Sports is third considering the Maden franchise and the like.
genre_descriptions = df2.groupby('Genre')
for genre,group in genre_descriptions:
    print(f"Descriptive statistics for genre: {genre}")
    print(group.describe())
    print("\n") 

grouped_sales = df2.groupby('Genre')["Global_Sales"].sum()
print(f"Total gloabl Sales by {grouped_sales}")

plt.figure(figsize=(10, 6))
grouped_sales.plot(kind='bar', color='skyblue')
plt.title('Total Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Global Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
#Grouping by Publisher, Nintendo dwarfs the market nearly twice over. Surprised Sony the maker of the playstion platforms is not higher.
publisher_descriptions = df2.groupby('Publisher')
for publisher,group in publisher_descriptions:
    print(f"Descriptive statistics for publisher: {publisher}")
    print(group.describe())
    print("\n") 

grouped_sales = df2.groupby('Publisher')["Global_Sales"].sum()
print(f"Total gloabl Sales by {grouped_sales}")

plt.figure(figsize=(6, 12))
grouped_sales.plot(kind='barh', color='skyblue')
plt.title('Total Global Sales by publisher')
plt.xlabel('Genre')
plt.ylabel('Total Global Sales')
plt.xticks(rotation=90)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
#Grouping by developer, many uinque vaues in this category. 
developer_descriptions = df2.groupby('Developer')
for developer,group in developer_descriptions:
    print(f"Descriptive statistics for developer: {developer}")
    print(group.describe())
    print("\n") 

grouped_sales = df2.groupby('Developer')["Global_Sales"].sum()
print(f"Total gloabl Sales by {grouped_sales}")

# I did not make a plot for each developer as there are early 500 unique developers. Wil reqire more EDA with something more.
print("Total amount of developers:",len(df2['Developer'].unique()))
```
